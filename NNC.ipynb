#convolución
usamos filtros para extraer caracterizticas
#Relu
No linealidad en imagenes, se aplica despues de cada covolución o a las capas de convoluciones.
#Max Pooling
Las caractersisticas relevantes son captadas de la imagen, evitando que la red neuronal solo aprenda algun determinado angulo o forma.
#Flattening
Aplanado de datos para la red neuronal
#full connection
Red Neuronal conectada

```
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
```

```
classifier=Sequential()
#convolution
classifier.add(Conv2D(filters=100,kernel_size=(4,4),
                      input_shape=(64,64,3),activation='relu'))#cuantos filtros filters, k
#max pooling
classifier.add(MaxPool2D(pool_size=(3,3)))
#convolution
classifier.add(Conv2D(filters=200,kernel_size=(3,3),
                      activation='relu'))
#max pooling
classifier.add(MaxPool2D(pool_size=(3,3)))
#convolution
classifier.add(Conv2D(filters=300,kernel_size=(2,2),
                      activation='relu'))
#max_pooling
classifier.add(MaxPool2D(pool_size=(2,2)))
#flattening aplanado
classifier.add(Flatten())
```

```
#full conect
classifier.add(Dense(units=56,activation='relu'))
classifier.add(Dropout(p=0.3))
classifier.add(Dense(units=28,activation='relu'))
classifier.add(Dropout(p=0.20))
classifier.add(Dense(units=4,activation='relu'))
classifier.add(Dropout(p=0.1))
classifier.add(Dense(units=1,activation='sigmoid'))
```

```
#compilando 
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
#generando imagenes
from keras.preprocessing.image import ImageDataGenerator
t_d=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,
                        horizontal_flip=True)
te_d=ImageDataGenerator(rescale=1./255)
t_ge=t_d.flow_from_directory('training_set',target_size=(64,64),batch_size=8,
                             class_mode='binary')#train
te_ge=te_d.flow_from_directory('test_set',target_size=(64,64),batch_size=8,
                               class_mode='binary')#test
```

```
classifier.fit_generator(t_ge,steps_per_epoch=8000,epochs=25,
                         validation_data=te_ge,validation_steps=100)
#steps_per_epoch:con cuantas imagenes entrena por epoch
#validation_steps: numero de pasos de validación, validamos 4000
```
